```python
import pandas as pd
import spacy
```

```python
#load spacy
nlp = spacy.load('en_core_web_sm')
df = pd.read_csv('cleaned_articles.csv')
```

```python
df.head()
```

```python
articles = df['articles']
```

```python
article1 = articles[0]
print(article1)
```

```python
def pos_tag(text):
    doc = nlp(text)
    return [(token.text, token.pos_) for token in doc]
```

```python
pos_tag(article1)
#visualise the pos tags using displacy

```

```python
from spacy import displacy
doc  = nlp(article1)
sentence_spans = list(doc.sents)
displacy.render(sentence_spans, style="dep", jupyter=True)
```

```python
df["pos_tags"] = df["articles"].apply(pos_tag)
df.to_csv("articles_pos.csv", index=False)
```

```python
def extract_ner(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]
```

```python
extract_ner(article1)
```

```python
doc = nlp(article1)
displacy.render(doc, style="ent", jupyter=True)
```

```python
df["ner_tags"] = df["articles"].apply(extract_ner)
df.to_csv("articles_ner.csv", index=False)
```
